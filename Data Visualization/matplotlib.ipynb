Matplotlib Basic Code

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

Supplots in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x**2,x**2)
```
